1. If Statements

```python
a = 10
b = 11

if a < b:
    print("a is less than b")
```

2. Elif Statements

```python
a = 12
b = 12
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

3. Else Statements

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

4. Short Hand Statement

```python
if a > b: print("a is greater than b")
```

5. Short Hand If..Else Statement

```python
a= 2
b = 330
print("A") if a > b else print("B")
```

6. And-both conditions are true

```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

7. Or-at least one of the conditions is true

```python
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
```

8. Nested If

```python
#Create a python program that will display above 10, above 20, above 30, above 40, but below 50
x = int(input("Enter a number: ")) #user input

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
    if x > 30:
      print("and also above 30!")
      if x > 40:
        print("and also above 40!")
        if x > 50:
          print("and also above 50!")

        else:
          print("but below 50.")
      else:
        print("but not above 40.")
    else:
      print("but not above 30.")
  else:
    print("but not above 20.")
else:
  print("but not above 10.")

# +5
```

# Supplementary Activity
Write a program to determine if the grades are:

# If:
     grade>=75, "Passed"
     grade = 74, "Remedial"
     grade < 74 "Failed"
     note: if grade is a negative number, it will not be accepted

```python
grade = int(input("Enter you grade: "))
if grade >= 75:
  print("Passed")
elif grade == 74:
  print("Remedial")
elif grade < 0:
  print("Invalid Grade")
else:
  print("Failed")

# +5
```
