Performing Operations with Python

```python
#1
print(value+value2)
```

```python
#2
value, value2 = 5.0, 6
#a. Addition
print(value+value2)

#Substraction
print(value-value2)
```

```python
#b.
value, value2 = -5, 6.1
print(value+value2)

#Substraction
print(value-value2)
```

```python
#c.
value, value2 = "Hello", ' world'
print(value+value2)

#Substraction for c. is not allowed because it is a string
```

```python
#d.
value, value2 = [1,2,3], [4,5,6]
print(value+value2)

#Substraction for d. is not allowed because it is a list
```

```python
#e.
value, value2 = (1,2,3), (4,5,6)
print(value+value2)

#Substraction for e. is not allowed because it is a tuple
```

```python
#f.
value, value2 = {"name":"Kenji"},{"age":"18"}
print(value,value2)

#Addition and Subtraction operators are not allowed because it is a dictionary
```
